```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/cleaned_survey_results_2019.csv')
df.head()
```

```python
df.columns.to_list()
```

```python
plt.figure(figsize=(10,6))
df['Which one of these is the closest to describing your role?'].dropna().str.split(';').explode().value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Roles')
plt.xlabel('Role')
plt.ylabel('Respondent Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Are you a freelancer/consultant?'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, figsize=(7,7),
                                         colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Respondent Type Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='What is your educational background', y='How many years of experience do you have doing professional data visualization?', data=df)
plt.xticks(rotation=45)
plt.title("Professional Visualization Experience by Educational Background")
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
numeric_cols_for_heatmap = [
    'How many years of experience do you have doing professional data visualization?',
    'Hours a week focused on data engineering?'
]

for col in numeric_cols_for_heatmap:
    df[col] = pd.to_numeric(df[col], errors='coerce')

plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_cols_for_heatmap].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
